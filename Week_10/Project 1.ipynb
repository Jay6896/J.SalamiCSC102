```python
import tkinter as tk

class zenith():
    def unique_services(self):
        self.userviceR = ["Retirement and education accounts", "Loans and mortgages", "Checking and saving"]
        self.userviceG = ["Multi-currency management services and products", "Foreign currency accounts", 
                          "Foreign currency credit cards", "Transborder advisory services", "Liquidity management"]
        self.userviceC = ["Advisory services"]

    def mutual_services(self):
        self.mservice = ["Lines of credit", "Investment management and accounts", "Insurance"]

class retail(zenith):
    def __init__(self):
        self.unique_services()
        self.mutual_services()

class globalBank(zenith):
    def __init__(self):
        self.unique_services()

class commercial(zenith):
    def __init__(self):
        self.unique_services()
        self.mutual_services()

class employee():
    def __init__(self, name, division):
        self.name = name
        self.division = division

def display_services():
    username = name_entry.get()
    userdiv = div_entry.get().upper()
    finalinput = employee(username, userdiv)

    output_text.delete(1.0, tk.END)  # Clear previous output

    if finalinput.division == "R":
        rvalues = retail()
        output_text.insert(tk.END, f"Welcome {username}, you are in Revision Banking\n")
        output_text.insert(tk.END, "You have 6 services:\n\n3 unique ones:\n")
        for service in rvalues.userviceR:
            output_text.insert(tk.END, f"• {service}\n")
        output_text.insert(tk.END, "\nand 3 mutual services with the others:\n")
        for service in rvalues.mservice:
            output_text.insert(tk.END, f"• {service}\n")
    elif finalinput.division == "G":
        gvalues = globalBank()
        output_text.insert(tk.END, f"Welcome {username}, you are in Global Banking\n")
        output_text.insert(tk.END, "You have 5 services:\n\n5 unique ones:\n")
        for service in gvalues.userviceG:
            output_text.insert(tk.END, f"• {service}\n")
        output_text.insert(tk.END, "\nand no mutual services with the others.\n")
    elif finalinput.division == "C":
        cvalues = commercial()
        output_text.insert(tk.END, f"Welcome {username}, you are in Commercial Banking\n")
        output_text.insert(tk.END, "You have 4 services:\n\n1 unique one:\n")
        for service in cvalues.userviceC:
            output_text.insert(tk.END, f"• {service}\n")
        output_text.insert(tk.END, "\nand 3 mutual services with the others:\n")
        for service in cvalues.mservice:
            output_text.insert(tk.END, f"• {service}\n")
    else:
        output_text.insert(tk.END, "Invalid division entered! Please enter R, G, or C.\n")

# Set up the main window
root = tk.Tk()
root.title("Zenith Bank Services")

# Create and place labels and entry fields for name and division
tk.Label(root, text="Enter your name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack(pady=5)

tk.Label(root, text="Enter your division (R, G, or C):").pack(pady=5)
div_entry = tk.Entry(root)
div_entry.pack(pady=5)

# Create a button to display services
btn_display = tk.Button(root, text="Display Services", command=display_services)
btn_display.pack(pady=10)

# Create a text area to show the results
output_text = tk.Text(root, height=15, width=50)
output_text.pack(pady=10)

# Start the Tkinter event loop
root.mainloop()
```
