```python
import pandas as pd
df = pd.read_csv('SIS.csv')

pirates = df[(df["Age"] > 14) & (df["Age"] < 18)]
yankees = df[(df["Age"] > 18) & (df["Age"] < 22)]
bulls = df[(df["Age"] > 22) & (df["Age"] < 25)]
aboveage = df[(df["Age"] > 25)]

class Students:
    def __init__(self, student, ineligble):
        self.student = student
        self.__ineligible = ineligble
        df

    def piratecsv(self, student):
        df = pd.DataFrame(student)
        df.to_csv(r'C:\Users\Jamil\Documents\J.SalamiCSC102\Week_12\The_Pirates.csv')
        print('The Pirates')
        for i in student:
            print(student[i])
        

    def yankeecsv(self, student):
        df = pd.DataFrame(student)
        df.to_csv(r'C:\Users\Jamil\Documents\J.SalamiCSC102\Week_12\The_Yankees.csv')
        print('The Yankees')
        for i in student:
            print(student[i])

    def bullcsv(self, student):
        df = pd.DataFrame(student)
        df.to_csv(r'C:\Users\Jamil\Documents\J.SalamiCSC102\Week_12\The_Bulls.csv')
        print('The Bulls')
        for i in student:
            print(student[i])

    def getter(self):
        print('Ineligible Users')
        for i in self.__ineligible:
            print(self.__ineligible[i])

userinp = Students(df, aboveage)
userinp.__init__
userinp.piratecsv(pirates)
userinp.yankeecsv(yankees)
userinp.bullcsv(bulls)
userinp.getter()


        
```
